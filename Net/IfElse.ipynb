```python
## Exercise: Shopping program, put 3 items inside the cart and the program will decides whether it passes the cost
## and quality standards. Standard: 500 > Cost > 100, Quality > 50
## Output: The shopping cart has <item 1>, <item 2>, and <item 3> which only <item 1>, <item 3> passes both cost
## and quality check while <item 2> failed and the total price is: <total price>

item1_name = str(input("Item 1 name: "))
item1_price = int(input("Item 1 Price: "))
item1_quality = int(input("Item 1 Quality: "))

item2_name = str(input("Item 2 name: "))
item2_price = int(input("Item 2 Price: "))
item2_quality = int(input("Item 2 Quality: "))

item3_name = str(input("Item 3 name: "))
item3_price = int(input("Item 3 Price: "))
item3_quality = int(input("Item 3 Quality: "))

sentence = "The shopping cart has " + item1_name + " " + item2_name + " " + item3_name + " " + " which only "

item1Quality = False
item1Price = False

item2Quality = False
item2Price = False

item3Quality = False
item3Price = False

# Checking the qualification

if item1_price > 100 and item1_price < 500:
  item1Price = True
  if item1_quality > 50:
    item1Quality = True

if item2_price > 100 and item2_price < 500:
  item2Price = True
  if item2_quality > 50:
    item2Quality = True

if item3_price > 100 and item3_price < 500:
  item3Price = True
  if item3_quality > 50:
    item3Quality = True

# Print out those that passes both check

if item1Price is True and item1Quality is True:
  sentence += item1_name + " "
if item2Price is True and item2Quality is True:
  sentence += item2_name + " "
if item3Price is True and item3Quality is True:
  sentence += item3_name + " "

sentence += " passes both cost and quality check while "

# Print out those that failed some check

if item1Price is False or item1Quality is False:
  sentence += item1_name + " "
if item2Price is False or item2Quality is False:
  sentence += item2_name + " "
if item3Price is False or item3Quality is False:
  sentence += item3_name + " "

totalPrice = item1_price + item2_price + item3_price

sentence += " failed and the total price is : " + str(totalPrice)

print(sentence)
```
