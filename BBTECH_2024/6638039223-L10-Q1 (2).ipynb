### Exercise 2: GPA Calculator

Write a function that receives a student's transcript as a **dictionary** and calculates the student's GPA in two decimal places. Each subject has a different weight, and each grade has a specific point value, as shown below:

| **Subject**   | **Weight** |
|---------------|------------|
| Calculus      | 4          |
| CompProg      | 4          |
| Physics       | 3          |
| Chemistry     | 3          |
| Biology       | 3          |
| Elective      | 2          |

| **Grade** | **Point Value** |
|-----------|------------------|
| A         | 4.0             |
| B+        | 3.5             |
| B         | 3.0             |
| C+        | 2.5             |
| C         | 2.0             |
| D+        | 1.5             |
| D         | 1.0             |
| F         | 0.0             |

The GPA should be calculated by taking into account the weight of each subject in the overall GPA formula.


````python
def gpa(transcript):
    # YOUR CODE HERE
    grade_points = {
        'A' : 4.0,
        'B+' : 3.5,
        'B' : 3.0,
        'C+' : 2.5,
        'C' : 2.0,
        'D+' : 1.5,
        'D' : 1.0,
        'F' : 0.0
    }
    subject_weights = {
        'Calculus' : 4,
        'CompProg' : 4,
        'Physics' : 3,
        'Chemistry' : 3,
        'Biology' : 3,
        'Elective' : 2
    }

    total_weighted_points = 0
    total_weights = 0

    for subject, grade in transcript.items():
        if grade in grade_points and subject in subject_weights:
            total_weighted_points += grade_points[grade] * subject_weights[subject]
            total_weights += subject_weights[subject]
    gpa_value = total_weighted_points / total_weights
    return round(gpa_value,2)
````

### Exercise 3: Student's Score

````python
def get_dict3(file_name) :
    # YOUR CODE HERE
    scores_dict = {}
    file = open(file_name,'r')
    for line in file:
        name, score = line.strip().split()
        scores_dict[name] = int(score)
    return scores_dict
def sorted_dict(s_dict) :
    # YOUR CODE HERE
    sorted_scores = dict(sorted(s_dict.items(),key=lambda item: item[1], reverse=True))
    return sorted_scores
````

### Exercise 4: Word Count
Write a function `get_dict(file_name)` that receives file name parameter for create dictionary from reading the text file. Then counts number of each words in content and packs in the dictionary.

**Example**
```
get_dict('conversation1.txt')
{'Tom': 2, 'Hi': 1, 'I': 2, 'am': 2, 'Ana': 2}
```

````python
def get_dict4(file_name):
    # YOUR CODE HERE
    word_count = {}
    file = open(file_name, 'r')
    for line in file:
        words = line.strip().split()
        for word in words:
            word = word.strip('.,!?"\':')
            if word in word_count:
                word_count[word] += 1
            else:
                word_count[word] = 1
    return word_count
````

### Exercise 5: Mahachulalongkorn
Write a function to reads the file **Mahachulalongkorn.txt** and creates a dictionary that counts how many times each word appears in the song. Then, input a word to check if it is in the song and display how many times it appears.  

**Mahachulalongkorn.txt**

```
namjai nong pee si chompu
tukkhon mairu luem boocha
phra khun khong laeng rean ma
chulalongkorn
koh thoon koh therd phra nam thai
phra khun nab wai ni run dorn
koh ong phra uea arthorn
lung porn krum krong

kwam dee tuk yang tang pong
pha yong phra geard krerk krai

cherd chai chayo
```

**Example**
```
word: chulalongkorn
The word chulalongkorn appears 1 times.
```

**Example**
```
word: koh
The word kho appears 4 times.
```

**Example**
```
word: chamchuri
The word chamchuri disappears.
```

````python
def mahachulalongkorn(filename,word):
    # YOUR CODE HERE
    word_count = {}
    file = open(filename, 'r')
    for line in file:
      words = line.strip().split()
      for w in words:
        w = w.strip('.,!?"\'')
        if w in word_count:
           word_count[w] += 1
        else:
           word_count[w] = 1
    word = word.lower()
    if word in word_count:
        return "The word " + word + " appears " + str(word_count[word]) + " times."
    else:
        return "The word " + word + " dissappears."
````
