### Exercise 1: Karyotype Analysis
Write a Python function `karyotype(fname)` that receives the name of a karyotype file (fname) and opens it to find abnormal pairs of chromosomes. The function should analyze the data to determine the potential disease based on the following conditions:
- **Edwards Syndrome:** The 18th chromosome pair has 3 chromosomes.
- **Down Syndrome:** The 21st chromosome pair has 3 chromosomes.
- **Klinefelter Syndrome:** The 23rd chromosome pair is "XXY".
- **Triple X Syndrome:** The 23rd chromosome pair is "XXX".
- **Jacob Syndrome:** The 23rd chromosome pair is "XYY".
- **Turner Syndrome:** The 23rd chromosome pair is "X" or "XO"

**Note that only one syndrome is reported based on the given order above.**

```python
def karyotype(fname):
    # YOUR CODE HERE
    f = open(fname, "r")
    F = f.readlines()
    numl = 0
    for line in F:
        line = line.strip()
        numl += 1
        if numl == 18 and len(line) == 3:
            return "Edwards syndrome"
        elif numl == 21 and len(line) == 3:
            return "Down Syndrome"
        elif numl == 23:
            if line == "XXY":
                return "Klinefelter Syndrome"
            elif line == "XXX":
                return "Triple X Syndrome"
            elif line == "XYY":
                return "Jacob Syndrome"
            elif line == "X" or line == "XO":
                return "Turner Syndrome"
```

**Test `karyotype(fname)`** <br>
Please run these cells below to check if your program works correctly.

```python
assert karyotype('karyotype1.txt') == 'Down Syndrome'
assert karyotype('karyotype2.txt') == 'Edwards Syndrome'
print("ok")
```

### Exercise 2: isPrime
Write a Python function `isPrime(number)` that receives a positive integer number and checks whether that number is a prime number. The function should return a boolean value indicating whether the number is prime.

```python
import math
def isPrime(number):
    # YOUR CODE HERE
    if number < 2:
        return False
    for i in range(2, int(math.sqrt(number))+1):
        if number % i == 0:
            return False
    return True
```

**Test `isPrime(number)`** <br>
Please run these cells below to check if your program works correctly.

```python
assert isPrime(3) == True
assert isPrime(97) == True
assert isPrime(2) == True
assert isPrime(17) == True
assert isPrime(101) == True
print("ok")
```

```python
assert isPrime(4) == False
assert isPrime(20) == False
assert isPrime(1) == False
assert isPrime(16) == False
print("ok")
```

### Exercise 3: Summation of Fibonacci Numbers
Write two Python functions:
- `fib(n)` that receives an integer number **n** and returns the $n^{th}$ fibonanci number.
- `sum_fib(n)` that receives an integer number **n** and returns the summation of the first **n** Fibonacci numbers using a function `fib(n)`

**Example:** if n=5, the summation is fib(1)+fib(2)+fib(3)+fib(4)+fib(5) = 1 + 1 + 2 + 3 + 5 = 12

```python
def fib(n):
    # YOUR CODE HERE
    if n == 1 or n == 2:
        return 1
    else:
        a,b = 1, 1
        for _ in range(n-2):
          a,b = b, a+b
          return b

def sum_fib(n):
    # YOUR CODE HERE
    total = 0
    for i in range(1, n+1):
        total += fib(i)
    return total
```

**Test `fib(n)` and `sum_fib(n)`** <br>
Please run these cells below to check if your program works correctly.

```python
assert fib(1) == 1
assert fib(2) == 1
assert fib(3) == 2
assert fib(4) == 3
assert fib(5) == 5
assert fib(6) == 8
assert fib(7) == 13
print("ok")
```

```python
assert sum_fib(4) == 7
assert sum_fib(5) == 12
assert sum_fib(6) == 20
assert sum_fib(7) == 33
assert sum_fib(8) == 54
print("ok")
```

### Exercise 4: Palindrome Checker
Write a Python function `palindrome(word)` that receives a number of words, then inputs those words and checks if each word is a palindrome.
The function should return a boolean value indicating whether the word is a palindrome.

```python
def palindrome(word):
    # YOUR CODE HERE
    length = len(word)
    for i in range(length//2):
        if word[i] != word[length-i-1]:
            return False
    return True
```

**Test `palindrome(word)`**<br>
Please run these cells below to check if your program works correctly.

```python
assert palindrome('madam') == True
assert palindrome('racecar') == True
assert palindrome('123321') == True
assert palindrome('selles') == True
print("ok")
```

```python
assert palindrome('python') == False
assert palindrome('word') == False
assert palindrome('0101') == False
assert palindrome('seller') == False
print("ok")
```
