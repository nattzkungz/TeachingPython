### Exercise 1: Products Stock
Write a function `get_dict(file_name)` that receives file name parameter and creates dictionary according to
- key is string of product code
- value is dictionary that have
    - key is string to definite size (`S`, `M`,  `L` and `XL`)
    - value is amount number of products

**Example**

```
get_dict('stock1.txt')
{4125: {'M': 4}, 5645: {'L': 7}, 7845: {'XL': 3}}
```

````python
def get_dict1(file_name):
    # YOUR CODE HERE
    product_dict = {}
    file = open(file_name, 'r')
    for line in file:
            parts = line.strip().split()
            product_code = int(parts[0].strip())
            size = parts[1].strip()
            quantity = int(parts[2].strip())
            if product_code not in product_dict:
                product_dict[product_code] = {}
            if  size in product_dict[product_code]:
                product_dict[product_code][size] += quantity
            else:
                product_dict[product_code][size] = quantity
    return product_dict
````

### Exercise 2: Caesar Cipher
Write a Python program that includes two functions: `caesar_encrypt(text, shift)` and `caesar_decrypt(text, shift)`. The first function should encrypt a given text by shifting each letter by a specified number **shift** in the alphabet, wrapping around when necessary. The second function should decrypt the encrypted text by reversing the shift.

**Example**
```
caesar_encrypt("Hello World", 3) # return: "Khoor Zruog"
caesar_decrypt("Khoor Zruog", 3) # return: "Hello World"
```

````python
def caesar_encrypt(text, shift):
    # YOUR CODE HERE
    encrypted_text = ""
    for char in text:
        if char.isalpha():
            ascii_offset = 65 if char.isupper() else 97
            encrypted_char = chr(((ord(char) - ascii_offset + shift)% 26) + ascii_offset)
            encrypted_text += encrypted_char
        else:
            encrypted_text += char
    return encrypted_text

def caesar_decrypt(text, shift):
    # YOUR CODE HERE
    decrypted_text = ""
    for char in text:
        if char.isalpha():
            ascii_offset = 65 if char.isupper() else 97
            decrypted_char = chr(((ord(char) - ascii_offset - shift) % 26) + ascii_offset)
            decrypted_text += decrypted_char
        else:
            decrypted_text += char
    return decrypted_text
````

### Exercise 4: Mendel's Pea Tree II
Write a function `Mendel(gemotype)` determines the phenotypes based on the genotype for a given number of genes by returning dictionary of phenotype.

**Example**
```
AaBBccDdffgg
{'Aa': 'Dominant', 'BB': 'Dominant', 'cc': 'Recessive', 'Dd': 'Dominant', 'ff': 'Recessive', 'gg': 'Recessive'}
```

````python
def mendel(genotype) :
    # YOUR CODE HERE
    phenotype_dict = {}
    for i in range(0, len(genotype), 2):
        gene_pair = genotype[i:i+2]

        if "A" in gene_pair or "B" in gene_pair or "C" in gene_pair or "D" in gene_pair or "E" in gene_pair or "F" in gene_pair or "G" in gene_pair:
            phenotype_dict[gene_pair] = "Dominant"
        else:
            phenotype_dict[gene_pair] = "Recessive"

    return phenotype_dict
````
