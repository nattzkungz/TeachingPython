##### Test cases

The unit test consists of
- the `test` function
- the test cases

To run the unit test
- run the definition of the `test` function
- call the `test` function with example inputs/outputs

If there is no `AssertionError`, the code is correct. Otherwise, you fail that test case. If you want to try more cases, you may add your own cases and try.

We have more hidden test cases that will be run later to grade your answer function.

```python
from unittest.mock import patch
from io import StringIO

def test_q1(inputs, outputs): 
    with patch('builtins.input', side_effect=inputs), patch('sys.stdout', new_callable=StringIO) as mock_stdout:
        answer_q1()
    
    soln = outputs.split(' ')
    sol1 = round(float(soln[-3]),5)
    sol2 = round(float(soln[-1]),5)
    sol_text = outputs[:4]
    
    print_text = mock_stdout.getvalue()
    ans = print_text.split(' ')
    ans1 = round(float(ans[-3]),5)
    ans2 = round(float(ans[-1]),5)
    ans_text = print_text[:4]
    
    assert sol_text == ans_text and sol1 == ans1 and sol2 == ans2, 'incorrect value: {}'.format(print_text)
```

```python
test_q1(['1,4,3'],'x = -1.0 , -3.0\n')
test_q1(['2.5,-8.7,2.1'],'x = 3.219053751558745 , 0.26094624844125425\n')

print('Congrats!. You have passed all of the test cases.')
```
