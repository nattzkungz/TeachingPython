# Q4 (6 pts.)

Riemann sum is an approximation of an integral by a finite sum. We will partition the region into rectagles and find the total area of these small partitions. This area is an approximation of a definite integral (area under the curve).

We are going to approximate the definite integral of a qudratic polynomial $f(x):a_2x^2+a_1x+a_0$ on a closed interval $[a,b]$. Firstly, we divide the interval $[a,b]$ into $p$ smaller subintervals of the same length. The area under the curve is a small rectangle whose base in the length of the subinterval and the height is $f(x)$ when $x$ is the midpoint of the subinterval.

Note the area is always positive even when $f(x)$ is negative.

Ref: https://en.wikipedia.org/wiki/Riemann_sum

```python
def answer_q4():
    # Your code here
    import math
    left,right = input("Enter left and right boundaries : ").split(",")
    left = float(left)
    right = float(right)
    
    partition = input("Enter number of partition : ")
    partition = float(partition)
    interval = (right-left)/partition
    mid = left+(interval/2)

    a2,a1,a0 = input("Enter qudratic polynomial (a2,a1,a0) : ").split(",")
    a2 = float(a2)
    a1 = float(a1)
    a0 = float(a0)

    riemann_sum = 0

    for num in range(int(partition)):
      p = ((mid**2)*(a2))+(mid*a1)+a0
      if (p<0): 
        p=-p
      
      riemann_sum = riemann_sum+(p*interval)
      mid +=interval
    print(riemann_sum)


    
```
